```python
N = 6  # Variant number
print("Ваш номер варіанту:", N)

# Використання модуля IPython для виведення елементів Markdown
from IPython.display import display, Markdown

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def task_6():
    a, b = map(int, input("Введіть два числа a і b через пробіл: ").split())
    primes = [x for x in range(a, b + 1) if is_prime(x)]
    print(primes)

task_6()

def task_66(a: int, b: int):
    if a > b:
        raise ValueError("Початкове значення a повинно бути меншим або рівним b.")
    primes = [x for x in range(a, b + 1) if is_prime(x)]
    return primes

# Тестові приклади
def run_tests():
    print("Тест 1: Вхідні дані: a=3, b=15")
    print("Результат:", task_66(3, 15))

    print("\nТест 2: Вхідні дані: a=6, b=24")
    print("Результат:", task_66(6, 24))

    print("\nТест 3: Вхідні дані: a=10, b=20")
    print("Результат:", task_66(10, 20))

run_tests()

display(Markdown("""
# Йорданія

<div style="border: 2px solid black; padding: 10px; margin: 5px;">
    <h2 style="color: #0033cc;">Загальна інформація</h2>
    <ul>
        <li><b>Столиця:</b> Амман</li>
        <li><b>Площа:</b> 89,342 км²</li>
        <li><b>Населення:</b> ~10,3 мільйонів</li>
        <li><b>Державний устрій:</b> <i>унітарна монархія</i></li>
        <li><b>Офіційна мова:</b> арабська</li>
        <li><b>Інші факти:</b>
            <ul>
                <li>Йорданія відома своєю багатою історією, включаючи стародавнє місто Петра.</li>
                <li>Мертве море — одна з найвідоміших природних пам'яток країни.</li>
            </ul>
        </li>
    </ul>
</div>

# Список видатних особистостей Йорданії
## Видатні особистості
- **Король Хусейн I** – впливовий лідер Йорданії, що сприяв стабільності у регіоні.
- **Марта Хігам** – археолог, що зробила внесок у дослідження Петри.

# Перелік визначних культурних та природних пам'яток Йорданії
## Пам'ятки культури та природи
1. **Петра** — стародавнє місто, одне з нових чудес світу.
2. **Мертве море** — найнижча точка на землі.
3. **Ваді-Рам** — пустеля з унікальними ландшафтами.

# Таблиця з цікавими фактами про різні міста Йорданії та її пам'ятками
### Таблиця цікавинок

| Місто       | Визначне місце              | Рік заснування |
|-------------|------------------------------|----------------|
| Амман       | Цитадель Аммана             | Стародавній Рим|
| Петра       | Археологічний парк Петра    | VI століття до н.е.|
| Акаба       | Червоне море                | Сучасність     |

# Гіперпосилання на офіційний туристичний портал Йорданії
### Гіперпосилання
[Офіційний туристичний портал Йорданії](https://www.visitjordan.com/)

# SVG графіка з зображенням прапора Йорданії
### Прапор Йорданії

<svg width="300" height="200">
  <rect width="300" height="66.7" fill="black"/>
  <rect y="66.7" width="300" height="66.7" fill="white"/>
  <rect y="133.4" width="300" height="66.7" fill="green"/>
  <polygon points="0,0 100,100 0,200" fill="red"/>
  <circle cx="50" cy="100" r="10" fill="white"/>
</svg>

# Зображення пам'яток Йорданії
### Фото Петри
![Петра](https://upload.wikimedia.org/wikipedia/commons/e/e7/Petra_Jordan_BW_36.JPG)

### Фото Мертвого моря
![Мертве море](https://upload.wikimedia.org/wikipedia/commons/d/d0/Dead_Sea_by_David_Shankbone.jpg)

"""))
```

```python
# Завдання 1
import pandas as pd

# Завантаження даних
df = pd.read_csv('NationalNames.csv')

# 1. Вивести перші 8 рядків набору даних
print(df.head(8))

# 2. Вивести останні 8 рядків набору даних
print(df.tail(8))

# 3. Отримати імена стовпців набору даних
print(df.columns.tolist())

# 5. Знайти кількість унікальних імен у наборі даних
print(df['Name'].nunique())

# 10. Підрахувати кількість унікальних імен у кожному році
unique_names_per_year = df.groupby('Year')['Name'].nunique()
print(unique_names_per_year)

# 11. Знайти рік із найбільшою кількістю унікальних імен
year_most_unique_names = unique_names_per_year.idxmax()
print(year_most_unique_names)

# 12. Знайти найпопулярніше ім’я в році з найбільшою кількістю унікальних імен
most_unique_year_data = df[df['Year'] == year_most_unique_names]
most_popular_name = most_unique_year_data.loc[most_unique_year_data['Count'].idxmax()]
print(most_popular_name)

# 13. Знайти рік, коли ім’я “Jacob” було найпопулярнішим серед жіночих імен
jacob_female = df[(df['Name'] == 'Jacob') & (df['Gender'] == 'F')]
if not jacob_female.empty:
    jacob_popular_year = jacob_female.loc[jacob_female['Count'].idxmax()]
    print(jacob_popular_year)
else:
    print("Ім’я Jacob не було популярним серед жіночих імен.")

# 14. Знайти рік із найбільшою кількістю гендерно нейтральних імен
gender_neutral = df.groupby(['Year', 'Name'])['Gender'].nunique()
gender_neutral_years = gender_neutral[gender_neutral == 2].reset_index()
neutral_counts = gender_neutral_years['Year'].value_counts()
print(neutral_counts.idxmax())

# 15. Знайти загальну кількість народжень за рік
total_births_per_year = df.groupby('Year')['Count'].sum()
print(total_births_per_year)

# 16. Знайти рік, коли народилося найбільше дітей
year_most_births = total_births_per_year.idxmax()
print(year_most_births)

# 17. Знайти кількість дівчаток та хлопчиків, які народились кожного року
births_by_gender = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()
print(births_by_gender)

# 18. Підрахувати кількість років, коли дівчаток народжувалось більше, ніж хлопчиків
girl_more_years = (births_by_gender['F'] > births_by_gender['M']).sum()
print(girl_more_years)

# 21. Порахуйте, скільки разів хлопчиків називали Barbara
barbara_male = df[(df['Name'] == 'Barbara') & (df['Gender'] == 'M')]['Count'].sum()
print(barbara_male)

# 22. Підрахуйте скільки років проводилось спостереження
years_observed = df['Year'].nunique()
print(years_observed)

# 23. Знати найпопулярніші гендерно-нейтральні імена (ті, що присутні кожного року)
popular_gender_neutral = gender_neutral_years['Name'].value_counts()
most_popular_neutral = popular_gender_neutral.idxmax()
print(most_popular_neutral)

# 24. Знайти найпопулярніше серед непопулярних імен
name_counts = df.groupby('Name')['Count'].sum()
least_popular = name_counts[name_counts == name_counts.min()]
most_popular_among_least = least_popular.idxmax()
print(most_popular_among_least)

# 26. Побудувати графіки розподілення кількості жіночих імен John та чоловічих імен Mary по роках
import matplotlib.pyplot as plt

# Дані для графіків
john_female = df[(df['Name'] == 'John') & (df['Gender'] == 'F')].groupby('Year')['Count'].sum()
mary_male = df[(df['Name'] == 'Mary') & (df['Gender'] == 'M')].groupby('Year')['Count'].sum()

# Графік для жіночих імен John
plt.figure(figsize=(10, 5))
plt.plot(john_female.index, john_female.values, label="Female John", marker='o')
plt.title("Кількість жіночих імен 'John' по роках")
plt.xlabel("Рік")
plt.ylabel("Кількість")
plt.legend()
plt.grid(True)
plt.show()

# Графік для чоловічих імен Mary
plt.figure(figsize=(10, 5))
plt.plot(mary_male.index, mary_male.values, label="Male Mary", marker='o', color='orange')
plt.title("Кількість чоловічих імен 'Mary' по роках")
plt.xlabel("Рік")
plt.ylabel("Кількість")
plt.legend()
plt.grid(True)
plt.show()
```
